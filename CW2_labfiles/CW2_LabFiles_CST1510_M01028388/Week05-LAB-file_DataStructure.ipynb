### Activity 1: 

Write a program that:
1. Creates a file called "my_profile.txt"
2. Writes 3 lines: your name, your age, and your favorite color
3. Reads the file back and prints each line with a label

**Example output:**
"Name: Alex", "Age: 20", "Favorite Color: Blue"
<hr>

`````python
#  Create and write information to the file
with open("my_profile.txt", "w") as file:
    file.write("Rudy\n")          # Name
    file.write("20\n")            # Age
    file.write("Black\n")          # Favorite color

# Read the file and print each line with a label
with open("my_profile.txt", "r") as file:
    lines = file.readlines()

# Remove any newline characters (\n)
name = lines[0].strip()
age = lines[1].strip()
color = lines[2].strip()

#  Display the formatted output
print("Name:", name)
print("Age:", age)
print("Favorite Color:", color)
`````

### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# Collect temperature input for 5 days and store them in a file
with open("temps.txt", "w") as file:
    for day in range(1, 6):
        temp = input(f"Enter temperature for Day {day}: ")
        file.write(f"Day {day}: {temp}°C\n")

# Read the stored data to process temperatures
with open("temps.txt", "r") as file:
    lines = file.readlines()

# Extract the temperature numbers and convert them to float values
temps = []
for line in lines:
    temp_value = line.split(":")[1].replace("°C", "").strip()
    temps.append(float(temp_value))

# Compute and print the average temperature
average_temp = sum(temps) / len(temps)
print("Average temperature over 5 days:", round(average_temp, 2), "°C")
`````

### **Activity 4: Movie Collection Tuple** 


Create a tuple with 3 movies and their release years:
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

Print:
- The oldest movie
- All movie titles
- Try to change a year (show it causes an error)

---

`````python
# create a tuple with movie names and their release years
movies = (("Inception", 2010), ("Matrix", 1999), ("Interstellar", 2014))

# find the oldest movie by comparing the years
oldest_movie = min(movies, key=lambda x: x[1])
print("Oldest movie:", oldest_movie[0], "released in", oldest_movie[1])

# print all movie titles
print("All movie titles:")
for movie in movies:
    print("-", movie[0])

`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
# define the hobbies for both friends
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

# we find what hobbies they both like
common_hobbies = alex_hobbies & sam_hobbies
print("Hobbies both enjoy:", common_hobbies)

# we find the hobbies that only Alex likes
alex_only = alex_hobbies - sam_hobbies
print("Hobbies only Alex enjoys:", alex_only)

# we combine all their hobbies together to see everything they like
all_hobbies = alex_hobbies | sam_hobbies
print("All unique hobbies combined:", all_hobbies)
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
# create a dictionary with students and their scores
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

# we add a new student with her score
grades["Diana"] = 88

# we update Bob's score to a new value
grades["Bob"] = 95

# we print students who scored more than 80
print("Students with scores above 80:")
for name, score in grades.items():
    if score > 80:
        print("-", name, ":", score)

# we calculate the average score of all students
average_score = sum(grades.values()) / len(grades)
print("Average score:", round(average_score, 2))
`````

### **Activity 7: Contact Book** 


Create a contact dictionary:
contacts = {
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

Write a program that:
- Adds a new contact
- Looks up a contact by name
- Prints all contacts in format "Name: Number"

---

`````python
# create a dictionary with some contacts
contacts = {
    "Mom": "555-1234",
    "Dad": "555-5678",
    "Friend": "555-9012"
}

# we add a new contact to the dictionary
name = input("Enter new contact name: ")
number = input("Enter their phone number: ")
contacts[name] = number

# we look up a contact by their name
search_name = input("Enter the name to look up: ")
if search_name in contacts:
    print(search_name + ":", contacts[search_name])
else:
    print("Contact not found.")

# we print all contacts nicely
print("\nAll contacts:")
for name, number in contacts.items():
    print(name + ":", number)
`````
