5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
#  We asssign marks of students in a list
marks = [78, 45, 89, 56, 32, 90, 67, 49, 73, 51]

#  Calculate average mark using a loop
total = 0
for mark in marks:
    total += mark
average = total / len(marks)

# Find highest and lowest mark
highest = max(marks)
lowest = min(marks)

# Count number of students who passed and failed
passed = 0
failed = 0

for mark in marks:
    if mark >= 50:
        passed += 1
    else:
        failed += 1

# Print results
print("Average mark:", round(average, 2))
print("Highest mark:", highest)
print("Lowest mark:", lowest)
print("Students passed:", passed)
print("Students failed:", failed)
````

12. **Dog Years**

It is commonly said that one human year is equivalent to 7 dog years. However, this simple conversion fails to recognise that dogs reach adulthood in approximately two years. As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. Ensure that your program works correctly for conversions of less than two or more humans’ years. Your program should display an appropriate error message if the user enters a negative number. 


````python
# Convert human years to dog years

human_years = float(input("Enter human years: "))

if human_years < 0:
    print("Error: Number of years cannot be negative.")
elif human_years <= 2:
    dog_years = human_years * 10.5
else:
    dog_years = 21 + (human_years - 2) * 4

print("The dog's age in dog years is:", round(dog_years, 2))
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random  # we use this to make a random number

# make a secret number between 1 and 20
secret = random.randint(1, 20)

# start try count at 0
tries = 0

# loop until user guesses right
while True:
    # ask user to guess a number
    guess = int(input("Guess the number (1-20): "))
    
    # add 1 to tries
    tries += 1

    # if user guess too small
    if guess < secret:
        print("Too small, try again")

    # if user guess too big
    elif guess > secret:
        print("Too big, try again")

    # if guess is right
    else:
        print("Nice! You got it!")
        break  # stop the loop

# show how many tries user needed
print("You needed", tries, "tries.")
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
# ask how many students
num_students = int(input("Enter number of students: "))

# make a list to keep all scores
scores = []

# get each student's score
for i in range(num_students):
    score = float(input("Enter student score: "))
    scores.append(score)

# check if we have enough scores
if len(scores) < 2:
    print("Need at least two scores to find highest and second highest.")
else:
    # sort scores from big to small
    scores.sort(reverse=True)

    # get first two biggest numbers
    highest = scores[0]
    second_highest = scores[1]

    # show results
    print("Highest score:", highest)
    print("Second highest score:", second_highest)
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
# We compares guess to real password, char-by-char (case + order matter)

actual = input("Enter the actual password: ")
guess = input("Enter the guess: ")

# we use length of actual as the base (extra chars in guess -> counted as wrong)
n = len(actual)

# count exact matches (same char, same position)
matches = 0
for i in range(n):
    # if guess too short, treat missing chars as mismatch
    if i < len(guess) and guess[i] == actual[i]:
        matches += 1

# compute accuracy as percent (rounded to nearest integer)
accuracy = round((matches / n) * 100) if n > 0 else 0

# show result
if accuracy == 100:
    print("Password cracked!")
else:
    print(f"Accuracy: {accuracy}%")
````
